```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pivottablejs import pivot_ui
import pandas_profiling as pdp
import seaborn as sns
from sklearn.datasets import load_boston
%matplotlib inline
```

```python
boston = load_boston()
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df
```

```python
plt.style.use('ggplot')
```

```python
fig = plt.figure(figsize=(16, 8))
ax1 = fig.add_subplot(1, 1, 1)
x_label = 'LSTAT'
y_label = 'CRIM'
ax1.plot(df[x_label], df[y_label], color='b', label=y_label)
ax1.scatter(df[x_label], df[y_label], color=(0.2,0.9,0.2), label=y_label)
#scatter for scatterplot and plot for line graph
ax1.set_xlabel(x_label)
ax1.set_ylabel(y_label)
ax1.set_title('graph')
fdict = {'size': 16}
ax1.legend(loc='upper left', bbox_to_anchor=(0, 0), borderaxespad=4, prop=fdict)
plt.show()
```
